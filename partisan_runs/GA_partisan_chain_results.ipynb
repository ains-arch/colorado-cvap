## Cut Edges

```python
import pickle
from gerrychain import Graph, GeographicPartition, Partition
```

```python
graph = pickle.load(open("../data/GA_graph_maupped_precincts.p", "rb"))
```

```python
enacted_cong_part = Partition(graph, assignment="CONDIST")
NUM_ENACTED_CONG_CUT_EDGES = len(enacted_cong_part["cut_edges"])
```

```python
cong_cut_edges = [cong_TOTPOP_cutedges, cong_VAP_cutedges,
                  cong_CPOP_cutedges, cong_CVAP_cutedges]
cong_cut_edge_min = min(list(map(min,cong_cut_edges)))
cong_cut_edge_max = max(list(map(max,cong_cut_edges)))
NUM_BINS = 15
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution of cut edges (Congressional Plans)")
plt.xlabel("Number of cut edges")
plt.hist(cong_cut_edges, alpha=0.5, bins=NUM_BINS,
          label=["TOTPOP", "VAP", "CPOP", "CVAP"])
plt.axvline(x=NUM_ENACTED_CONG_CUT_EDGES, color="k", linestyle="dashed",
            label="Enacted plan")
plt.legend()
plt.show()
```

```python
enacted_senate_part = Partition(graph, assignment="SENDIST")
NUM_ENACTED_SENATE_CUT_EDGES = len(enacted_senate_part["cut_edges"])
```

```python
senate_cut_edges = [senate_TOTPOP_cutedges, senate_VAP_cutedges,
                    senate_CPOP_cutedges, senate_CVAP_cutedges]
senate_cut_edge_min = min(list(map(min,senate_cut_edges)))
senate_cut_edge_max = max(list(map(max,senate_cut_edges)))
NUM_BINS = 15
```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution of cut edges (Senate Plans)")
plt.xlabel("Number of cut edges")
plt.hist(senate_cut_edges, alpha=0.5, bins=NUM_BINS,
          label=["TOTPOP", "VAP", "CPOP", "CVAP"])
plt.axvline(x=NUM_ENACTED_SENATE_CUT_EDGES, color="k", linestyle="dashed",
            label="Enacted plan")
plt.legend()
plt.show()
```
